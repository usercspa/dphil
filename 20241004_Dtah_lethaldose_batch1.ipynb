```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the table
days_post_irradiation = [7, 14, 21, 27, 32, 43, 57, 64]
dose_gy = [0, 10, 20, 30, 40, 80, 100]
data = np.array([
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 7, 8, 8, 7, 8],
    [7, 5, 7, 8, 3, 3, 2],
    [7, 3, 4, 8, 0, 1, 0],
    [7, 0, 1, 1, 0, 0, 0],
    [7, 0, 0, 0, 0, 0, 0]
])

# Reference values provided for each dose
reference = np.array([7, 8, 8, 8, 9, 8, 9])

# Survival rate calculation using the provided reference values for each dose
# Repeat the reference values across all days to match the shape of the data
survival_rate = data / reference

# Plotting
plt.figure(figsize=(10, 6))

for i, dose in enumerate(dose_gy):
    plt.plot(days_post_irradiation, survival_rate[:, i], label=f'Dose {dose} Gy', marker='o')

# Add labels and title
plt.xlabel('Days Post Irradiation')
plt.ylabel('Survival Rate (Relative to Provided Reference)')
plt.title('Survival Rate Over Time by Radiation Dose')

# Show legend outside the plot
plt.legend(title='Dose (Gy)', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the table
days_post_irradiation = [7, 14, 21, 27, 32, 43, 57, 64]
dose_gy = [0, 10, 20, 30, 40, 80, 100]
data = np.array([
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 7, 8, 8, 7, 8],
    [7, 5, 7, 8, 3, 3, 2],
    [7, 3, 4, 8, 0, 1, 0],
    [7, 0, 1, 1, 0, 0, 0],
    [7, 0, 0, 0, 0, 0, 0]
])

# Reference values provided for each dose
reference = np.array([7, 8, 8, 8, 9, 8, 9])

# Survival rate calculation using the provided reference values for each dose
survival_rate = data / reference

# Colorblind-friendly palette (Okabe & Ito)
cb_colors = ["#000000", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00"]
# Distinct point shapes
markers = ['o', 's', '^', 'D', 'P', 'X', '*']

plt.figure(figsize=(10, 6))

for i, (dose, color, marker) in enumerate(zip(dose_gy, cb_colors, markers)):
    plt.errorbar(
        days_post_irradiation,
        survival_rate[:, i],
        label=f'{dose} Gy',
        color=color,
        marker=marker,
        linestyle='-',
        markersize=8,
        linewidth=2,
        capsize=4
    )

plt.xlabel('Days Post Irradiation', fontsize=14, fontweight='bold', family='sans-serif')
plt.ylabel('Survival Rate (relative to reference)', fontsize=14, fontweight='bold', family='sans-serif')
plt.title('Survival Rate Over Time by Radiation Dose', fontsize=16, fontweight='bold', family='sans-serif')
plt.xticks(days_post_irradiation, fontsize=12, family='sans-serif')
plt.yticks(fontsize=12, family='sans-serif')
plt.ylim(-0.05, 1.15)
plt.grid(axis='y', linestyle=':', color='grey', alpha=0.5)
plt.legend(title='Dose (Gy)', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```
