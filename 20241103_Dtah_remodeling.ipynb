```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your data
df = pd.read_csv('/content/20231223_DT_remodeling_rate.csv')

# Calculate the Pre-to-Post ratio if not already done
def calculate_ratio(df):
    if 'Pre-pharyngeal_length' in df.columns and 'Post-pharyngeal_length' in df.columns:
        df['Post-pharyngeal_length'] = df['Post-pharyngeal_length'].replace(0, np.nan)
        df['Pre_to_Post_ratio'] = df['Pre-pharyngeal_length'] / df['Post-pharyngeal_length']
    else:
        print("Required columns are not present in the DataFrame.")
    return df

df = calculate_ratio(df)

# Separate the uncut data as control and the anterior fragment data for analysis
uncut_df = df[df['Fragment'] == 'uncut']
anterior_df = df[df['Fragment'] == 'anterior']

# Calculate average and standard error for each species and time point for both anterior and uncut
summary_anterior = anterior_df.groupby(['Time_(dpa)', 'Species']).agg(
    average_ratio=('Pre_to_Post_ratio', 'mean'),
    sem_ratio=('Pre_to_Post_ratio', lambda x: x.std() / np.sqrt(len(x)))
).reset_index()

summary_uncut = uncut_df.groupby(['Time_(dpa)', 'Species']).agg(
    average_ratio=('Pre_to_Post_ratio', 'mean'),
    sem_ratio=('Pre_to_Post_ratio', lambda x: x.std() / np.sqrt(len(x)))
).reset_index()

# Calculate the mean ratio for each species in the uncut control
species_uncut_means = uncut_df.groupby('Species')['Pre_to_Post_ratio'].mean()

# Define shapes for each species
shapes = {
    'Smed': 'v',   # Triangle for Smed
    'Dtah': 's'    # Square for Dtah
}

# Plotting
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Plot anterior fragment data with error bars
for species in summary_anterior['Species'].unique():
    subset = summary_anterior[summary_anterior['Species'] == species]
    plt.errorbar(subset['Time_(dpa)'], subset['average_ratio'],
                 yerr=subset['sem_ratio'],
                 fmt=shapes[species],
                 label=f'{species} (Anterior)',
                 capsize=5,
                 elinewidth=2,
                 markersize=8)

# Plot uncut control data with the same shape but a dashed line style
for species in summary_uncut['Species'].unique():
    subset = summary_uncut[summary_uncut['Species'] == species]
    plt.errorbar(subset['Time_(dpa)'], subset['average_ratio'],
                 yerr=subset['sem_ratio'],
                 fmt=shapes[species],
                 label=f'{species} (Uncut Control)',
                 capsize=5,
                 elinewidth=2,
                 markersize=8,
                 linestyle='--') # Dashed line for control

# Add a dashed horizontal line at the uncut control mean for each species
for species, mean_ratio in species_uncut_means.items():
    plt.axhline(y=mean_ratio, color='gray', linestyle='--', linewidth=1.5, label=f'{species} Uncut Control Mean')

# Customize the plot
plt.title('Remodeling of Anterior Fragment Over Time with Uncut Control')
plt.xlabel('Day post amputation')
plt.ylabel('Pre-to-Post Pharyngeal Length Ratio')
plt.legend(title='Species and Fragment Type')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('/content/20231223_DT_remodeling_rate - Cleaned.csv')

# Ensure numeric columns are float
for col in ['Pre-pharyngeal_length', 'Post-pharyngeal_length']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calculate ratio
df['ratio'] = df['Pre-pharyngeal_length'] / df['Post-pharyngeal_length']

# Get baseline (uncut) mean ratio for each species
baseline = df[df['Fragment'] == 'uncut'].groupby('Species')['ratio'].mean().to_dict() #Calculate Baseline (Uncut) Ratio for Each Species

# Normalize ratio to uncut baseline for each species
df['ratio_norm'] = df.apply(lambda row: row['ratio'] / baseline[row['Species']], axis=1) #This divides each sample's ratio by the mean uncut ratio for its species, so all data is relative to the uncut baseline (uncut = 1).

# Aggregate: mean and SD for each group
summary = (
    df[df['Fragment'] != 'uncut']
    .groupby(['Time', 'Species', 'Fragment'])
    .agg(mean_ratio_norm=('ratio_norm', 'mean'),
         sd_ratio_norm=('ratio_norm', 'std'))
    .reset_index()
)

# For plotting, sort Time as integer
summary['Time'] = summary['Time'].astype(int)
summary = summary.sort_values('Time')

# Colorblind-friendly palette and marker/linestyle mapping
palette = {'Smed': '#0072B2', 'Dtah': '#E69F00'}
linestyles = {'anterior': 'solid', 'posterior': 'dashed'}
markers = {'anterior': 'o', 'posterior': 's'}

plt.figure(figsize=(8, 6))
for (species, fragment), group in summary.groupby(['Species', 'Fragment']):
    plt.errorbar(
        group['Time'], group['mean_ratio_norm'], yerr=group['sd_ratio_norm'],
        label=f'{species} {fragment}',
        color=palette[species],
        linestyle=linestyles[fragment],
        marker=markers[fragment],
        capsize=4
    )

plt.axhline(1, color='grey', linestyle=':', lw=1)  # baseline
plt.xlabel('Days post-amputation')
plt.ylabel('Normalized pre/post-pharyngeal length ratio\n(relative to uncut)')
plt.title('Remodeling dynamics in planarian fragments\n(pre/post-pharyngeal ratio, normalized)')
plt.legend(title='Group', loc='best')
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# If not already loaded:
# df = pd.read_csv('your_data.csv')  # Or however you loaded your data

# Clean up column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Ensure numeric columns are float
for col in ['Pre-pharyngeal_length', 'Post-pharyngeal_length']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calculate ratio
df['ratio'] = df['Pre-pharyngeal_length'] / df['Post-pharyngeal_length']

# Get baseline (uncut) mean ratio for each species
baseline = df[df['Fragment'] == 'uncut'].groupby('Species')['ratio'].mean().to_dict()

# Normalize ratio to uncut baseline for each species
df['ratio_norm'] = df.apply(lambda row: row['ratio'] / baseline[row['Species']], axis=1)

# Function to plot for a given fragment type
def plot_fragment(fragment_type, ax):
    data = (
        df[df['Fragment'] == fragment_type]
        .groupby(['Time', 'Species'])
        .agg(mean_ratio_norm=('ratio_norm', 'mean'),
             sd_ratio_norm=('ratio_norm', 'std'))
        .reset_index()
    )
    data['Time'] = data['Time'].astype(int)
    data = data.sort_values('Time')
    palette = {'Smed': '#0072B2', 'Dtah': '#E69F00'}
    markers = {'Smed': 'o', 'Dtah': 's'}
    for species, group in data.groupby('Species'):
        ax.errorbar(
            group['Time'], group['mean_ratio_norm'], yerr=group['sd_ratio_norm'],
            label=species,
            color=palette[species],
            marker=markers[species],
            linestyle='-',
            capsize=4
        )
    ax.axhline(1, color='grey', linestyle=':', lw=1)
    ax.set_xlabel('Days post-amputation')
    ax.set_ylabel('Normalized pre/post-pharyngeal ratio\n(relative to uncut)')
    ax.set_title(f'{fragment_type.capitalize()} fragments')
    ax.legend(title='Species')
    sns.despine(ax=ax)

# Create the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
plot_fragment('anterior', axes[0])
plot_fragment('posterior', axes[1])
plt.suptitle('Remodeling dynamics in planarian fragments\n(pre/post-pharyngeal ratio, normalized)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
