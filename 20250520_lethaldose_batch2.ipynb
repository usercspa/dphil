```python
# --- Step 1: Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

df = pd.read_csv('/content/LethaldoseDT2.csv', header=None)

# --- Step 4: Extract Relevant Data ---

# Days post-irradiation: Row 13 (index 12), columns C to U (index 2 to 20)
days_post_irradiation = df.iloc[12, 2:21].astype(float).tolist()

# Dose in Gy: Column B (index 1), rows 14 to 20 (index 13 to 19)
dose_gy = df.iloc[13:20, 1].str.extract(r'(\d+)').astype(float)[0].tolist()

# Survival data: Rows 14 to 20 (index 13 to 19), columns C to U (index 2 to 20)
raw_data = df.iloc[13:20, 2:21]

# Convert percentage strings to float (e.g., '100.0%' -> 1.0)
survival_rate = raw_data.replace('%', '', regex=True).astype(float) / 100

# --- Step 5: Plotting ---

# Colorblind-friendly palette (Okabe & Ito)
cb_colors = ["#000000", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00"]
markers = ['o', 's', '^', 'D', 'P', 'X', '*']

plt.figure(figsize=(10, 6))

for i, (dose, color, marker) in enumerate(zip(dose_gy, cb_colors, markers)):
    plt.plot(
        days_post_irradiation,
        survival_rate.iloc[i],
        label=f'{int(dose)} Gy',
        color=color,
        marker=marker,
        linestyle='-',
        markersize=8,
        linewidth=2
    )

plt.xlabel('Days Post Irradiation', fontsize=14, fontweight='bold', family='sans-serif')
plt.ylabel('Survival Rate', fontsize=14, fontweight='bold', family='sans-serif')
plt.title('Survival Rate Over Time by Radiation Dose', fontsize=16, fontweight='bold', family='sans-serif')
plt.xticks(days_post_irradiation, fontsize=12, family='sans-serif')
plt.yticks(fontsize=12, family='sans-serif')
plt.ylim(-0.05, 1.05)
plt.grid(axis='y', linestyle=':', color='grey', alpha=0.5)
plt.legend(title='Dose (Gy)', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# --- Step 1: Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

df = pd.read_csv('/content/LethaldoseDT2.csv', header=None)

# --- Step 4: Extract Full Dataset ---

# Days post-irradiation: row 13 (index 12), columns C to U (index 2 to 20)
days = df.iloc[12, 2:21].astype(float)
weeks = (days / 7).round(1)  # Convert to weeks with 1 decimal precision

# Dose in Gy: column B (index 1), rows 14 to 20 (index 13 to 19)
dose_gy = df.iloc[13:20, 1].str.extract(r'(\d+)').astype(float)[0].tolist()

# Survival data: rows 14 to 20, columns C to U
raw_data = df.iloc[13:20, 2:21]
survival_rate = raw_data.replace('%', '', regex=True).astype(float) / 100

# --- Step 5: Plotting ---

# Colorblind-friendly palette (Okabe & Ito)
cb_colors = ["#000000", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00"]
markers = ['o', 's', '^', 'D', 'P', 'X', '*']

plt.figure(figsize=(10, 6))

for i, (dose, color, marker) in enumerate(zip(dose_gy, cb_colors, markers)):
    plt.plot(
        weeks,
        survival_rate.iloc[i],
        label=f'{int(dose)} Gy',
        color=color,
        marker=marker,
        linestyle='-',
        markersize=8,
        linewidth=2
    )

plt.xlabel('Weeks Post Irradiation', fontsize=14, fontweight='bold', family='sans-serif')
plt.ylabel('Survival Rate', fontsize=14, fontweight='bold', family='sans-serif')
plt.title('Survival Rate Over Time by Radiation Dose (Weeks)', fontsize=16, fontweight='bold', family='sans-serif')

# Set xticks based on rounded weeks
plt.xticks(np.arange(0, weeks.max() + 1, 1), fontsize=12, family='sans-serif')
plt.yticks(fontsize=12, family='sans-serif')
plt.ylim(-0.05, 1.05)
plt.grid(axis='y', linestyle=':', color='grey', alpha=0.5)
plt.legend(title='Dose (Gy)', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# --- Step 1: Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Step 2: Data from the table ---
days_post_irradiation = [7, 14, 21, 27, 32, 43, 57, 64]
dose_gy = [0, 10, 20, 30, 40, 80, 100]

data = np.array([
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 8, 8, 9, 8, 9],
    [7, 8, 7, 8, 8, 7, 8],
    [7, 5, 7, 8, 3, 3, 2],
    [7, 3, 4, 8, 0, 1, 0],
    [7, 0, 1, 1, 0, 0, 0],
    [7, 0, 0, 0, 0, 0, 0]
])

# Reference values provided for each dose
reference = np.array([7, 8, 8, 8, 9, 8, 9])

# --- Step 3: Calculate relative survival rate ---
survival_rate = data / reference

# --- Step 4: Convert time axis to weeks ---
days = np.array(days_post_irradiation)
weeks = np.round(days / 7, 1)  # Convert to weeks, rounded to 1 decimal

# --- Step 5: Plotting ---
cb_colors = ["#000000", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00"]
markers = ['o', 's', '^', 'D', 'P', 'X', '*']

plt.figure(figsize=(10, 6))

for i, (dose, color, marker) in enumerate(zip(dose_gy, cb_colors, markers)):
    plt.plot(
        weeks,
        survival_rate[:, i],
        label=f'{int(dose)} Gy',
        color=color,
        marker=marker,
        linestyle='-',
        markersize=8,
        linewidth=2
    )

plt.xlabel('Weeks Post Irradiation', fontsize=14, fontweight='bold', family='sans-serif')
plt.ylabel('Survival Rate (relative to baseline)', fontsize=14, fontweight='bold', family='sans-serif')
plt.title('Relative Survival Rate Over Time by Radiation Dose', fontsize=16, fontweight='bold', family='sans-serif')

# Standardized x-axis ticks (every 1 week)
plt.xticks(np.arange(0, np.max(weeks) + 1, 1), fontsize=12, family='sans-serif')
plt.yticks(fontsize=12, family='sans-serif')
plt.ylim(-0.05, 1.15)
plt.grid(axis='y', linestyle=':', color='grey', alpha=0.5)
plt.legend(title='Dose (Gy)', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```
